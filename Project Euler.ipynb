# Problem 4:

#### A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 Ã— 99.

#### Find the largest palindrome made from the product of two 3-digit numbers.

```python
def find_palindrome(num):
    if len(str(num))%2==0:
        split = int(len(str(num))/2)
        first = str(num)[:split]
        end = str(num)[split:]
        rev_end = end[::-1]
        
        if first==rev_end:
            return True
            #print('Num = {} is palindrome'.format(num))
        else:
            return False
            print('Num = {} is not palindrome'.format(num))
    else:
        split = int(len(str(num))//2)
        first = str(num)[:split]
        end = str(num)[split+1:]
        rev_end = end[::-1]
        
        if first==rev_end:
            return True
            #print('Num = {} is palindrome'.format(num))
    
        else:
            return False
            #print('Num = {} is not palindrome'.format(num))
```

# Problem 5 
#### 2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.
#### What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
def smallest_divisor(n):
    '''Function that finds smallest positive real that is divisible by first n integers'''
    
    #initialize a and b
    a = n
    b = 2
    
    # while b is less than n
    while b<=n:
        # if b is a factor of a - then a remains and b increments up 1 (to n)
        if a%b==0:
            a=a
            b+=1
        # Else increment a up one and return b to 2
        else:
            a+=1
            b=2
    return a

ans = smallest_divisor(20)
```

#### By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.
#### What is the 10 001st prime number?

```python
def find_prime(n):
    # initialize list
    prime_list=[]
    
    # search range from 2 to n
    for i in range(2, n+1):
        # divide each number from 2 up to i into i - if any number up to i is a factor - number is not prime
        for j in range(2,i):
            if i%j==0:
                break
        # Number is prime - add to list
        else:
            prime_list.append(i)
    return prime_list
find_prime(548)
```

# Problem 9:
#### There exists exactly one Pythagorean triplet for which a + b + c = 1000.
#### Find the product abc.

```python
# pythagorean triplet is a set of numbers such that a < b < c and a^2 + b^2 = c^2
# EX 3^2 + 4^2 = 5^2

# we know that a + b + c = 1000
# min of a = 1
# min of b = 2
# max of c = 997 ( if a=1 and b=2)
def pythagorean_triplet(n):
    '''find the pythagorean triplet whose sum of numbers is n'''
    # initialize a b and c such that a is less than b and b is less than c
    for a in range(1, n+1):
        for b in range(a+1, n+1):
            c = n - a - b
            if a**2 + b**2 == c**2:
                return a*b*c, a,b,c

pythagorean_triplet(1000)
```

Find_primes works but at O(~n^2) because each number n has to be checked for each value from 2 to n to determine whether or not it is prime.  This is too slow for large numbers of n.

```python
 def primeSieve(n):
    # Returns a list of prime numbers calculated using the Sieve of Eratosthenes algorithm.
    
    # Create list of TRUE values - of length n
    sieve = [True] * n
    # zero and one are not prime number
    sieve[0] = False
    sieve[1] = False

    # create the sieve - 
    # only need to check up to sqrt(n) because if n is NOT prime then n=a*b and at most a or b can equal sqrt(n)
    for i in range(2, int(np.sqrt(n)) + 1):
        # 
        #print(i)
        pointer = i * 2
        while pointer < n:
            #print(pointer)
            # sets values to false that are multiples of 2, 3, 4, 5,... to sqrt(n)
            sieve[pointer] = False
            pointer += i
        #print('done with pointer', pointer)
    # compile the list of primes
    primes = []
    for i in range(n):
        if sieve[i] == True:
            primes.append(i)

    return primes
```
