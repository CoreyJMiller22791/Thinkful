```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline

data = pd.read_csv('Pokemon.csv', index_col=0)
df = pd.DataFrame(data)

#print(df.head())
#print(df.info())
#print(df.nunique())
#print(df['Type 1'].value_counts())
```

For this challenge I will include HP, Attack, Special Attack, Defense, Special Defense, and Speed as the continuous variables and Type 1 as the categorical variable.

Let's only consider the pokemon from the first generation.

Goal: create one plot that shows probability distributions, bivariate relationships, and variance of relationships across groups

```python
#start by looking at pdf for each var
dropcols = ['Name', 'Type 2', 'Total', 'Legendary', 'Generation']
dfcon = df[df.columns.drop(dropcols)].reset_index(drop=True)

dropfunc = []
def masterplot(*columns):
    """create function to remove continuous variables and categories from plot"""
    for column in columns:
        dropfunc.append(column)
    temp = dfcon[dfcon.columns.drop(dropfunc)]  
    sns.pairplot(temp, hue='Type 1')
    plt.show()

#masterplot('HP', "Attack")
masterplot()
```

What we see above is the 18 different types of Type 1 plotted in a 6x6 matrix of the continuous variables, where each variable is compared to each other and itself.  The diagonal plots contain the pdf for each of the continuous variables.  The other plots show the bivariate relationships (or lack thereof) for each variable against another.
