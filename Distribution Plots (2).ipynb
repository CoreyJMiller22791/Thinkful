# Distribution Plots

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# normal distribution with mean=10 and stdev=2 
norm = np.random.normal(10, 2, 100)
plt.hist(norm, color='c')
plt.title('Normal Distribution \n mean=10 stdev=2\n n=100')
plt.axvline(norm.mean(), color='m', linewidth=3)
plt.axvline(norm.mean()+norm.std(), color='k', linestyle='dashed')
plt.axvline(norm.mean()-norm.std(), color='k', linestyle='dashed')
plt.show()
print("Mean: {}".format(norm.mean()))
print("Stdev: {}".format(norm.std()))
```

Normal distribution separates the data such that 50% of the data lies to the left of the mean and the other 50% lies to the right. The mean is the center of the normal distribution and when symmetric the mean=median=mode. The standard deviation for a normal distribution will tell us information about the dispersion or variability in the data when compared to the mean. 1 standard deviation from the mean will include roughly 68% of the data values, 2 will contain 95% and 3 99.7%.

```python
# Poisson distribution with lambda=5
poiss = np.random.poisson(5, 100)
plt.hist(poiss, color='b')
plt.title('Poisson Distribution \n lambda=5 \n n=100')
plt.axvline(poiss.mean(), color='r', linewidth=3)
plt.axvline(poiss.mean()+poiss.std(), color='y', linestyle='dashed', linewidth=3)
plt.axvline(poiss.mean()-poiss.std(), color='y', linestyle='dashed', linewidth=3)
plt.show()
print("Mean: {}".format(poiss.mean()))
print("Stdev: {}".format(poiss.std()))
print(5**0.5)
```

Poisson distributions are specified by one parameter, lambda.  Lambda is equal to the mean and the variance.  As lambda increases the distribution approaches normal.  Poisson is used to describe the number of times an event occurs in a finite observation space.  A variable follows this distribution when data are counts of an event (non-negative and unbounded), independent, and the average rate does not change of the period of observation.  This distribution is similar to binomial in that they both count events, but the major difference is that binomial places an upper bound on the number of events equal to the number of times the test is run.

```python
# Gamma distribution with shape alpha=12 and scale beta=4
gam = np.random.gamma(12, 4, 100)
plt.hist(gam, color='r')
plt.title('Gamma Distribution \n shape alpha=12 scale beta=4 \n n=100')
plt.axvline(gam.mean(), color='k', linewidth=3)
plt.axvline(gam.mean()+gam.std(), color='g', linestyle='dashed', linewidth=2)
plt.axvline(gam.mean()-gam.std(), color='g', linestyle='dashed', linewidth=2)
plt.show()
print("Mean: {}".format(gam.mean()))
print("Stdev: {}".format(gam.std()))
print((12*4**2)**0.5)
```

Gamma distribution takes two parameters, alpha and beta.  The product of these two parameters is equal to the mean.  The variance is alpha*beta^2 and stdev is equal to the sqrt(var). Gamma distributions are commonly used with processes where the waiting time between events are relevant.

```python
# Binomial distribution with # of trials=10, probability of success=.7, and tested 100 times
binom = np.random.binomial(10, .7, 100)
plt.hist(binom, color='y')
plt.title('Binomial Distribution \n trials=10 probability=0.7\n number of tests=100')
plt.axvline(binom.mean(), color='k', linewidth=3)
plt.axvline(binom.mean()+binom.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binom.mean()-binom.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
print("Mean: {}".format(binom.mean()))
print("Stdev: {}".format(binom.std()))
print((10*.7*.3)**0.5)
```

The binomial distribution has two parameters, the number of trials and the probability of success.  The product of these two parameters is the mean and the sqrt(np(1-p)) is the stdev.  Binomial distributions measure the number of successes in a certain observation space with the number of events being at most equal to the number of tests that are run.  This is the key difference between binomial and poisson distributions, Poisson distributions are based on continuous events.

```python
# Chisquared distribution with 4 degrees of freedom and n=100
chi = np.random.chisquare(4, 100)
plt.hist(chi, color='k')
plt.title('Chi Square Distribution \n df=4 \n n=100')
plt.axvline(chi.mean(), color='c', linewidth=3)
plt.axvline(chi.mean()+chi.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(chi.mean()-chi.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
print("Mean: {}".format(chi.mean()))
print("Stdev: {}".format(chi.std()))
print((2*4)**0.5)
```

Chi-squared distribution has one parameter, the degrees of freedom.  The degrees of freedom is equal to the mean.  The variance for Chi-squared is 2*degrees of freedom; therefore, the standard deviation is the sqrt(2dof).  Chi-squared distributions are positively skewed with the degree of skew decreasing with increasing degrees of freedom. Thus, we see as degree of freedom increases we approach normal in the Chi-squared distribution.

```python
# Rayleigh distribution with scale=3 and n=100
ray = np.random.rayleigh(3, 100)
plt.hist(ray, color='m')
plt.title('Rayleigh Distribution \n scale=3 \n n=100')
plt.axvline(ray.mean(), color='g', linewidth=3)
plt.axvline(ray.mean()+ray.std(), color='k', linestyle='dashed', linewidth=2)
plt.axvline(ray.mean()-ray.std(), color='k', linestyle='dashed', linewidth=2)
plt.show()
print('Mean: {}'.format(ray.mean()))
print('Stdev: {}'.format(ray.std()))
print(3*(3.14/2)**0.5)
print((3*(2-(3.14/2))**0.5))
```

The rayleigh distribution has one parameter, the shape.  The mean of the rayleigh distribution is equal to shape*sqrt(pi/2).  The standard deviation is shape(sqrt(2-(pi/2))).  This distribution is unlike the rest in the sense that characteristics of the distribution and the parameters don't explicitly tell us the mean or standard deviation like the other distributions.

# Additional normal distributions

```python
norm1 = np.random.normal(5, 0.5, 100)
norm2 = np.random.normal(10, 1, 100)
norm3 = norm1 + norm2

plt.hist(norm3)
plt.axvline(norm3.mean(), color='k', linewidth=3)
plt.axvline(norm3.mean()+norm3.std(), color='y', linestyle='dashed', linewidth=3)
plt.axvline(norm3.mean()-norm3.std(), color='y', linestyle='dashed', linewidth=3)
plt.show()

print('Mean: {}'.format(norm3.mean()))
print('Stdev: {}'.format(norm3.std()))
print((1**2+.5**2)**0.5)
```

The sum of the two random normal distributions has a mean equal to the sum of the two independent means.  The variance is also the sum of the individual variances; therefore, the standard deviation is the sqrt(var1+var2).
